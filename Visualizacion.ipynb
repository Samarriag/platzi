# **SubplotS** POO

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=np.sin(x)
```



1.   Crear figura
2.   Entrar al axes



```python
fig, axes=plt.subplots()
axes.plot(x,y, 'b')
```

```python
fig, axes=plt.subplots(nrows=1, ncols=2)
#axes[0] (axes sub cero es un array)
#haciendo referencia a los arrays, entrar de ellos atraves del indice
axes[0].plot(x,y, 'b')
axes[1].plot(y,x, 'r')
```

```python
#si ya se que los axes son tipo array, se podria definir desde el inicio como una tupla
fig, (ax1,ax2)=plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y, 'b')
ax2.plot(y,x, 'r')
```

```python
# subplot(filas,columnas)
fig, axes=plt.subplots(2, 4)
axes[0,0].plot(x,np.cos(x), 'y')
axes[0,1].plot(x,np.sin(x), 'r')
axes[0,2].plot(x,np.tan(x))
axes[0,3].plot(x,np.cos(x)**2)

axes[1,0].plot(x,np.cos(x), 'y')
axes[1,1].plot(x,np.sin(x), 'r')
axes[1,2].plot(x,np.tan(x))
axes[1,3].plot(x,np.cos(x)**2)
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8))=plt.subplots(2, 4)
ax1.plot(x,np.cos(x), 'y')
ax2.plot(x,np.sin(x), 'r')
ax3.plot(x,np.tan(x))
ax4.plot(x,np.cos(x)**2)
fig.tight_layout()
```

#Leyendas, etiquetas, titulos, tamano

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

Darle contexto a las graficas

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('Relation x - y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[1].plot(y,x)
axes[1].set_title('Relation y - x')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

Aumentar tamano de las graficas y agregar labels

```python
fig, axes = plt.subplots(1,2, figsize=(10,5))
axes[0].plot(x,y, label='sin(x)')
axes[0].set_title('Relation x - y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relation y - x')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

Metodo POO

```python
plt.plot(x, y, label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left', bbox_to_anchor=(0.1,0.1))
```

# Otros graficos con Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Crear datos aleatorios de 1 a 50, 100 registros
data = np.random.randint(1,50,100)
```

```python
data
```

# Histograma

```python
plt.hist(data, bins=10, width=5, histtype='bar')
#bins=numero de barras, width=ancho de barras, histtype=tipo de histograma
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step')
plt.show()
```

#Boxplot

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data, vert=False, patch_artist=True,notch= True)
#vert=False -horizontal- , patch_artist=True -rellenar caja-, notch= True -senala datos hacia la mediana-
plt.show()
```

```python
#Si hay OUTLIERS
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True,notch= True)
#vert=False -horizontal- , patch_artist=True -rellenar caja-, notch= True -senala datos hacia la mediana-
plt.show()
```

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True,notch= True, showfliers=False)
#showfliers=False - NO MUESTRA OUTLIERS-
plt.show()
```

#Jointplot & Pairplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', 
              #obtener mas valor de los graficos secundarios con marginal
              #para obtener tabla para la escala de los histogramas
              marginal_ticks=True,
              #keywords= diccionario (no. bins (barras), multiple es separano sexo )
              marginal_kws=dict(bins=25, fill=True, multiple='dodge'))
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=False)
plt.show()
```

```python
sns.pairplot(data=tips)
plt.show()
      
```

```python
#corner para no tener information duplicada 
sns.pairplot(data=tips,hue='sex',palette='coolwarm',corner=True,kind='scatter')
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex");
```
